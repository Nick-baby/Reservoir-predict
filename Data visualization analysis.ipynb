```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats 
import os
currentPath = os.getcwd().replace('\\', '/')
source=pd.read_excel(currentPath+'/Source_data/por/N1-N2-N3.xlsx')
target=pd.read_csv(currentPath+'/target_data/Well logged data/A1.csv')
```

```python
Label_source="Source"
label_target="A1"     
```

```python
from scipy.stats import *
plt.figure()
sns.distplot(source['GR'].values, kde=True,bins=50,color="#ef4026",kde_kws={"label":Label_source,'linestyle':"solid"})
sns.distplot(target["GR_1"].values, kde=True, bins=50, color="#0339f8",kde_kws={"label":label_target,'linestyle':"-."})
plt.legend(loc="upper right",fontsize=15)
plt.xticks(fontsize=23)
plt.yticks(fontsize=23)
plt.xlabel('GR',fontsize=23)
plt.ylabel("Density",fontsize=23)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.manifold import TSNE
tsne = TSNE(n_components=1, init='pca', random_state=0)
src = tsne.fit_transform(source.values[:,1:-2])
tar = tsne.fit_transform(target.values[:,1:-1])
sns.distplot(src, kde=True,bins=50,color="#ef4026",kde_kws={"label":"Source",'linestyle':"solid"} )
sns.distplot(tar, kde=True,bins=50,color="#0339f8",kde_kws={"label":"A1",'linestyle':"-."} )
plt.legend(loc="upper right",fontsize=15)
plt.xticks(fontsize=23)
plt.yticks(fontsize=23)
plt.ylabel("Density",fontsize=23,)
plt.xlabel("T-SNE Map Value",fontsize=23)
plt.show()
```
